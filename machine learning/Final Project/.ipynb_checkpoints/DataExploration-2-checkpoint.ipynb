```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

## Loading the data

```python
#load the full csv
#full = pd.read_csv('C:/Users/jchun/OneDrive/Documents/425/fulldata.csv')
full = pd.read_csv('fulldata.csv')
```

```python
del full['Unnamed: 0']   #delete"Unnamed: 0"
del full['onlinedating.1']  #delete"onlinedating.1"
```

```python
full.head() #show the data after cleaning
```

## Removing columns that have values of 0

```python
#delete all the variables with 0 in order to get the data with meaning
for col in full.columns:
    if(0 in full[col].values):
        del full[col]
```

```python
#add the recession feature("rec"), since there is a delay when the job market react to the recession, we let the recession feature in 2008 to 2011 equals to 1  
#full['rec']
full['rec'] = full.apply(lambda row: 1 if row['Year'] in [2008,2009,2010,2011] else 0,axis=1)
#full['rec'] = full.apply(lambda row: 1 if row['Year'] > 2007 else 0,axis=1)
```

```python
#After removing the variables with 0, there are 23 variables left. They are shown in the following. 1th variable: Year; 
#2rd: State; 3th: Fiscal Policy ; 4th: Regulatory Policy; 5th: Personal Freedom to do jobs; 6th:Economic Freedom to do choice
#7th: Overall freedom; 8th: unemployment rate; 9th: beer sale; 10th: wine sale; 11th: how many times people go to church; 
#12th: 64 bit operating system; 13th: how to measure whether pregnant or not; 14th: jobsin network webite for searching jobs
#15th: anxiety; 16th: onlinedating apps; 17th: racism; 18th: racism words; 19th: online class; 20th:jokes; 21th: football game
#22th: religion; 23th:recession background
full.columns
```

```python
#to see the relationship between features and ground truth label(unemployment rate), we plot a correlation plot. From the result, we can see that 64bit, jobsin and rec are positively related to the unemployment rate.
from matplotlib import pyplot

dims = (20, 20)
fig, ax = pyplot.subplots(figsize=dims)

#cols = [c for c in df.columns if (c.lower()[:8] != 'country_' or \
#        c == 'country_US') and 'os' not in c and 'screen' not in c and 'browser' not in c]
cols = [c for c in full.columns if (c.lower()[:8] != 'country_')]
dfCorr=full[cols]

corr = dfCorr.corr()
#Pastel1 terrain
sns.heatmap(ax=ax,data=corr,cmap="terrain")
```

```python
# We visualized our ground truth label, unemployment rate. 
plt.hist(x=full['unemployment'], bins='auto', color='#0504aa',alpha=0.7, rwidth=0.85)
```

```python
#We classified the unemployment rate into 0(unemployment rate smaller than 7) and 1(unemployment rate bigger than 7)
full['highunem'] = full.apply(lambda row: 1 if row['unemployment'] > 7 else 0, axis=1)
```

```python
full.head()
```

```python
#divide the full sample into training data and testing data
from random import shuffle
n = len(full) #number of rows in your dataset
indices = list(range(n))
shuffle(indices)

df = full.loc[indices[:400]]
dfTest = full.loc[indices[400:]]
```

```python
df.columns
```

```python
#we will use these variables in our model
indFeats = [
            'beer','wine',
    '64bit','pregnancy','jobsin','anxiety','onlinedating','racism','nigger',
    'onlineclass','jokes','football','religion','church','rec']
    #'Fiscal.Policy','Regulatory.Policy','Personal.Freedom','Economic.Freedom','Overall.Freedom']
```

```python
#normalize the features before regressing them because our label are in (0,1) form
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Create a features DF
from sklearn.preprocessing import normalize

dfX = df[indFeats]
df_test_X = dfTest[indFeats]
dfX = normalize(dfX, axis=0, norm='max')
df_test_X = normalize(df_test_X, axis=0, norm='max')

# Create a label DF
dfY = df['unemployment']
df_test_Y = dfTest['unemployment']
```

## Linear Regression Model

```python
lm = linear_model.LinearRegression()
lm.fit(dfX, dfY)

dfPred = lm.predict(df_test_X)
```

```python
import statsmodels.api as sm

model = sm.OLS(dfY,dfX)
results = model.fit()
print(results.summary())
```

```python
print(indFeats)
# The coefficients
print('Coefficients: ', lm.coef_)
print('Intercept: ', lm.intercept_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(np.array(df_test_Y),dfPred))
print("Mean absolute error: %.2f" % mean_absolute_error(np.array(df_test_Y),dfPred))
```

After getting the regression results, we calculate the MSE and MAE of testing samples for our first model so that we can compare the accuracy of the  model with other models.

```python
import matplotlib.pyplot as plt

dims = (10, 10)
fig, ax = pyplot.subplots(figsize=dims)
plt.title('Coefficients for all Features')
g = sns.barplot(ax=ax, x=indFeats, y=lm.coef_).set_xticklabels(rotation=90,labels=indFeats)
```

## Random Forest

```python
df.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score


dfBoolY = df['highunem']
dt = DecisionTreeClassifier(max_depth=6,max_leaf_nodes=12) #set the maximum number of the layer to 6
print(cross_val_score(dt, dfX, dfBoolY, cv=10))
mod = dt.fit(dfX,dfBoolY,)
```

```python
#visualize the decision tree
from sklearn import tree
import graphviz

dot_data = tree.export_graphviz(mod, out_file=None, 
                         feature_names=indFeats,  
                         class_names='isgritty',  
                         filled=True, rounded=True,  
                         special_characters=True)  
graph = graphviz.Source(dot_data)  
graph
```

```python
#random forest regression
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=10,max_features=8,n_estimators=100)
rf.fit(dfX, dfY)

dims = (10, 10)
fig, ax = pyplot.subplots(figsize=dims)
plt.title('Coefficients for all Features Predicting Unemployment')
g = sns.barplot(ax=ax, x=indFeats, y=rf.feature_importances_).set_xticklabels(rotation=90,labels=indFeats)

print(indFeats)
print(rf.feature_importances_)
```

```python
#Mean Square Error of the random forest model
dfPred = rf.predict(df_test_X)
print("Mean squared error: %.2f" % mean_squared_error(np.array(df_test_Y),dfPred))
print("Mean absolute error: %.2f" % mean_absolute_error(np.array(df_test_Y),dfPred))
```

## Neural Network

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
# define the model
def larger_model():
    # create model
    model = Sequential()
    model.add(Dense(20, input_dim=15, kernel_initializer='normal', activation='relu'))
    model.add(Dense(16, kernel_initializer='normal', activation='relu'))
    model.add(Dense(8, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    # Compile model
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
```

```python
seed = 10
np.random.seed(seed)
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=larger_model, epochs=50, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(pipeline, dfX, dfY, cv=kfold)
print("Larger: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```

```python
model = larger_model()
model.fit(dfX, dfY, epochs=1000, verbose=0)
```

```python
#Mean Square Error of the neural network model
dfPred = model.predict(df_test_X)
print("Mean squared error: %.2f" % mean_squared_error(np.array(df_test_Y),dfPred))
print("Mean absolute error: %.2f" % mean_absolute_error(np.array(df_test_Y),dfPred))
```
